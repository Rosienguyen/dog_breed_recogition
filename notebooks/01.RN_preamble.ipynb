```python
import os
import random
import shutil

# set the path to the original dataset
data_dir = '../archive/images/images'

# set the ratio for the train, validation and test set
train_ratio = 0.8
val_ratio = 0.1
test_ratio = 0.1

# create the three directories for train, val and test
train_dir = os.path.join('../data/train')
os.makedirs(train_dir)
val_dir = os.path.join('../data/val')
os.makedirs(val_dir)
test_dir = os.path.join('../data/test')
os.makedirs(test_dir)
```

```python


# loop through each sub-folder (breed) in the original dataset
for breed in os.listdir(data_dir):
    breed_dir = os.path.join(data_dir, breed)
    if os.path.isdir(breed_dir):
        # create the breed directory in the three splits
        train_breed_dir = os.path.join(train_dir, breed)
        os.makedirs(train_breed_dir)
        val_breed_dir = os.path.join(val_dir, breed)
        os.makedirs(val_breed_dir)
        test_breed_dir = os.path.join(test_dir, breed)
        os.makedirs(test_breed_dir)

        # get the list of images in this breed folder
        images = os.listdir(breed_dir)
        # shuffle it
        random.shuffle(images)
        # divide the images into train/val/test according to the predefined ratios
        num_train = int(train_ratio * len(images))
        num_val = int(val_ratio * len(images))
        train_images = images[:num_train]
        val_images = images[num_train:(num_train+num_val)]
        test_images = images[(num_train+num_val):]

        # move the image files into the appropriate subdirectory
        for img in train_images:
            src = os.path.join(breed_dir, img)
            dst = os.path.join(train_breed_dir, img)
            shutil.copyfile(src, dst)
        for img in val_images:
            src = os.path.join(breed_dir, img)
            dst = os.path.join(val_breed_dir, img)
            shutil.copyfile(src, dst)
        for img in test_images:
            src = os.path.join(breed_dir, img)
            dst = os.path.join(test_breed_dir, img)
            shutil.copyfile(src, dst)
```
